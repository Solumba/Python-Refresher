```python
import networkx as nx
import time

class UAVNetworkSimulation:

    def create_initial_graph(self, num_uavs=10):
        G = nx.Graph()
        for i in range(num_uavs):
            # Initialize with random values for demonstration
            G.add_node(i, latency=10, throughput=100)
        return G

    def run_simulation(self, G, total_time, update_interval):
        start_time = time.time()
        iteration = 0
        while time.time() - start_time < total_time:
            print(f"\nIteration {iteration}:")
            # Directly modifying properties for each node
            for node_id, attr in G.nodes(data=True):
                # Update latency and throughput
                attr['latency'] += 5  # Increment latency by 5
                attr['throughput'] -= 10  # Decrement throughput by 10

                # Ensure throughput doesn't drop below zero
                if attr['throughput'] < 0:
                    attr['throughput'] = 0

                # Print current node state
                print(f"Node {node_id}: Latency = {attr['latency']}, Throughput = {attr['throughput']}")

            iteration += 1
            time.sleep(update_interval)


# Usage
simulation = UAVNetworkSimulation()
G = simulation.create_initial_graph(num_uavs=5)
simulation.run_simulation(G, total_time=20, update_interval=1)
```
